# Optical Fiber 
An optical fiber is a thin fiber of glass or plastic that can carry light from one end to the other. 
The study of optical fibers is called fiber optics, which is part of applied science and engineering.


Optical fibers are mainly used in telecommunications, but they are also used for lighting, sensors, toys, and special cameras for seeing inside small spaces. It is sometimes used in medicine to see inside people, like down their throat.
## Problem statement
Determine whether or not specified materials will form a light pipe.
If they do form a light pipe, compute the angles at which light can
enter the pipe and be transmitted.

### Input / Output description
As shown in Table Index of refraction the input to the program is a
data file containing the indices of refraction for the potential light
pipes. The output is a message indicating whether or not light is
transmitted and the angles at which it can enter the pipe.


#### Index of refraction for common materials

| Acrylic | 1.49 | 

### Simple Example
The index of refraction of air is 1.0003 and the index of glass is
1.5. If we from a light pipe of glass surrounded by air, the critical
angle c can be computed as follows:

$$c = sin^{-1}(n2/n1)= sin^{-1}( 1.0003/1.5)= sin^{-1}(0.66687)= 41.82^{o}$$

This light pipe will transmit light for all angles of incidence greater
than 41.82o

###  python solution

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

file = open('opticalfiber.dat','r')# we import the file opticalfiber.dat and read it
Data = file.readlines()
file.close()
```

```python
x = []
y = []

for i in Data[1:]:
        c0 = i.split('\t')[0] #column 1 with names
        c1 = i.split('\t')[1] #column 2 with values
        x.append(c0)        
        y.append(float(c1))
```

```python
#lets check the imported data
for i in range(len(x)):
    print(x[i], y[i])
```

```python
for i in range(len(x)):
    for j in range(i+1, len(y)):
        if(i == j):
            continue
        print('Index of refraction {0} = {1} and index of refraction {2} = {3}'.format(x[i], y[i], x[j], y[j]))
        if (y[i] > y[j]):
            print('The pipe will not transmit light!','\n') 
            continue
        else:
            radian = np.arcsin((y[i]) / (y[j]))  
            angle = round(np.degrees(radian), 2)  #get the angle in degrees
            print('Angle  of incidence = {0}'.format( angle))
            print('The pipe will transmit light!','\n') 
```
